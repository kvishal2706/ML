```python
import numpy as np
import tensorflow as tf
```

```python
def load_data():
    X=np.load("data/X.npy")
    y=np.load("data/y.npy")
    X_train=X[0:1000]
    y_train=y[0:1000]
    return X_train,y_train
```

```python
X_train,y_train=load_data()

print(X_train.shape)
print(y_train.shape)
```

```python
model=tf.keras.models.Sequential([
    tf.keras.Input(shape=(400,)),
    tf.keras.layers.Dense(units=25,activation="sigmoid"),
    tf.keras.layers.Dense(units=15,activation="sigmoid"),
    tf.keras.layers.Dense(units=1,activation="sigmoid"),
])
```

```python
model.summary()
```

```python
[layer1,layer2,layer3]=model.layers

w1,b1=layer1.get_weights()
w2,b2=layer2.get_weights()
w3,b3=layer3.get_weights()

print(w1.shape,b1.shape)
print(w2.shape,b2.shape)
print(w3.shape,b3.shape)
```

```python
model.compile(loss=tf.keras.losses.binary_crossentropy,
              optimizer=tf.keras.optimizers.Adam(0.001),)  #which loss function to use here
model.fit(X_train,y_train,
          epochs=100)      #minimise w and b and choose correct value of alpha(basically gradient descent)
```

```python
correct=0

m=X_train.shape[0]
yhat=100

for i in range(m):
    prediction=model.predict(X_train[i].reshape(1,400))
    if prediction >= 0.5:
        yhat = 1
    else:
        yhat = 0

    if yhat == y_train[i]:
        correct+=1

print(correct)
```
