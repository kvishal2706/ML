House Training Example

```python
import numpy as np
import matplotlib.pyplot as plt
# plt.style.use('./deeplearning.mplstyle')
```

```python
x_train = np.array([1.0, 1.7, 2.0, 2.5, 3.0, 3.2])  # sizes in 1000 sq ft
y_train = np.array([250, 300, 480,  430,   630, 730,])#prices in 1000s of dollars
print(f"x_train : {x_train}")
print(f"y_train : {y_train}")
```

```python
print(f"x_train.shape : {x_train.shape}")  # will give the tuple entry for the dimension here 1 dimension that has 2 elements 
m=x_train.shape[0]      # m is the no of training examples
print(f"no of training ex : {m}")  # will give the tuple entry for the dimension
```

```python
plt.scatter(x_train,y_train, marker='x',c="r")  # plot the graph
plt.title('Housing Prices')
plt.ylabel("price in (1000s of dollars)")
plt.xlabel("size in (1000 sqft)")

plt.show() #to display the data
```

```python
w=200
b=100
print(f"w: {w}")
print(f"b: {b}")
```

```python
def compute_model_output(x,w,b):
    m=x.shape[0]
    f_wb=np.zeros(m)
    for i in range(m):
        f_wb[i]=w*x[i]+b

    return f_wb
```

```python
tmp_f_wb=compute_model_output(x_train,w,b)
print(tmp_f_wb)

plt.plot(x_train,tmp_f_wb,c="b",label="Our Prediction")
plt.scatter(x_train,y_train,marker='x',c="r",label="Actual Values")
plt.title("Housing Price")
plt.ylabel("price in (1000s of dollars)")
plt.xlabel("size in (1000 sqft)")
plt.legend()
plt.show()
```
