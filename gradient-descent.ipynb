```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0, 1.7, 2.0, 2.5, 3.0, 3.2])  # sizes in 1000 sq f
y_train = np.array([250, 300, 480,  430,   630, 730,])#prices in 1000s of dollars
print(f"x_train: {x_train}")
print(f"y_train: {y_train}")
```

```python
plt.scatter(x_train,y_train,c="r",marker="x")
plt.xlabel("size in (1000sq f)")
plt.ylabel("price in (1000s $)")
plt.title("House Price Prediction")
# plt.legend()
plt.show()
```

```python
def compute_cost(x,y,w,b):
    m=x.shape[0]
    f_wb=np.zeros(m)
    for i in range(m):
        f_wb[i]=w*x[i]+b

    return f_wb
```

```python
def gradient_diff(x,y,w,b):
    m=x.shape[0]
    dj_dw=0
    dj_db=0
    for i in range(m):
        f_wb=w*x[i]+b
        dj_dw_i=(f_wb-y[i])*x[i]
        dj_db_i=(f_wb-y[i])
        dj_dw = dj_dw + dj_dw_i
        dj_db = dj_db + dj_db_i
    dj_dw=dj_dw/m
    dj_db=dj_db/m

    return dj_dw,dj_db
```

```python
def gradient_descent(x,y,alpha,w,b,iteration):
    for i in range(iteration):
        dj_dw,dj_db=gradient_diff(x,y,w,b)
        w=w-alpha*dj_dw
        b=b-alpha*dj_db
    
    return w,b
```

```python
w_in=0
b_in=0
alpha=0.01
iteration=1000

tmp_fwb=compute_cost(x_train,y_train,w_in,b_in)

plt.plot(x_train,tmp_fwb,c="b",label="Prev Prediction")
plt.scatter(x_train,y_train,marker="x",c="r")
plt.xlabel("size in (1000sq f)")
plt.ylabel("price in (1000s $)")
plt.title("House Price Prediction")
plt.legend()
plt.show()
```

```python
w_final,b_final=gradient_descent(x_train,y_train,alpha,w_in,b_in,iteration)
print(w_final)
print(b_final)

f_wb_final=compute_cost(x_train,y_train,w_final,b_final)

plt.plot(x_train,f_wb_final,c="b",label="Prev Prediction")
plt.scatter(x_train,y_train,marker="x",c="r",label="Actual Values")
plt.xlabel("size in (1000sq f)")
plt.ylabel("price in (1000s $)")
plt.title("House Price Prediction")
plt.legend()
plt.show()

```
